In this dataset you have to predict that whether we have to give ATTORNEY to the clients or not (0=Yes, 1=No) based on below factors.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot
import seaborn as sns
```

```python
df  = pd.read_csv("claimants.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df['CLMSEX'] = df['CLMSEX'].fillna(df['CLMSEX'].mode()[0])
df['CLMINSUR'] = df['CLMINSUR'].fillna(df['CLMINSUR'].mode()[0])
df['SEATBELT'] = df['SEATBELT'].fillna(df['SEATBELT'].mode()[0])
df['CLMAGE'] = df['CLMAGE'].fillna(df['CLMAGE'].mean())
```

```python
df.head()
```

```python
df.isna().sum()
```

## Splitting the data

```python
x = df.drop('ATTORNEY',axis = 1)
y = df['ATTORNEY']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=123456)
```

```python
from sklearn.linear_model import LogisticRegressionCV
model = LogisticRegressionCV()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

### confusion matrix, Calculate accuracy score, precision score, recall score, f1 score, roc auc score.

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix,RocCurveDisplay
cm = confusion_matrix(y_test,y_pred)
accuracy = accuracy_score(y_test,y_pred)
precision = precision_score(y_test,y_pred)
recall = recall_score(y_test,y_pred)
f1 = f1_score(y_test,y_pred)
# ROC = RocCurveDisplay(y_test,y_pred)
```

```python
print(f"accuracy score: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"recall: {recall:.2f}")
print(f"f1 score: {f1:.2f}")
print(f"Confusion matrix: {cm}")
```
