The delivery_time.csv dataset contains information about Delivery time and Sorting Time. We have to prepare linear regression and polynimial regression model to predict Sorting time on the basis of Delivery Time. Also evaluate the model by using evaluation metrics.

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("delivery_time.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.heatmap(df.corr(),annot=True)
```

### splitting data

```python
x = df.drop('Sorting Time',axis = 1)
y = df['Sorting Time']
```

```python
df.isna().sum()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,test_size=0.2,random_state=123456)
```

## building linear regression model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

# Model Evaluation 

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error,root_mean_squared_error,r2_score

y_pred = model.predict(x_test)
MAE = mean_absolute_error(y_test,y_pred)
MSE = mean_squared_error(y_test,y_pred)
RMSE = root_mean_squared_error(y_test,y_pred)
R2 = r2_score(y_test,y_pred)
print(f"MAE = {MAE:.2f}, MSE = {MSE:.2f}, RMSE = {RMSE:.2f}, R2 = {R2:.2f}")
```

```python
predictions = model.predict(x)
for i in range(0,len(list(x['Delivery Time']))):
    print(f"Sorting time for delivery time {x['Delivery Time'][i]:.2f} is {predictions[i]:.2f}")
```
