```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("NewspaperData.csv")
df.head()
```

## Data cleansing

```python
# Here we dont need the newspaper column , so we'll drop it
df.drop("Newspaper",axis = 1, inplace = True)
```

# EDA

```python
# checking corr()
df.corr()
```

```python
df.isna().sum()
```

## Splitting the data

```python
x = df.drop('sunday',axis = 1)
y = df['sunday']
type(x)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=123456)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error,r2_score
MAE = mean_absolute_error(y_test,y_pred)
MSE = mean_squared_error(y_test,y_pred)
RMSE = root_mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
r2Adjusted = 1 - (1-r2**2)*((len(y_pred)+1)/(len(y_pred)-(len(x_test)+1)))
```

```python
predictions = model.predict(x)
for i in range(0,len(list(x['daily']))):
    print(f"Daily sale: {x['daily'][i]:.2f} Sunday sales: {predictions[i]:.2f}")
```
