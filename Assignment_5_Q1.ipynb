```python
for i in df.columns:
    corr = df['Type'].corr(df[i])
    if corr >= -0.1 and corr <= 0.1:
        df.drop(i,axis=1,inplace = True)
```

```python
df.head()
```

```python
x = df.drop('Type',axis = 1)
y = df['Type']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=123456)
```

```python
def Decision_tree():
    from sklearn.tree import DecisionTreeClassifier
    model = DecisionTreeClassifier()
    model.fit(x_train,y_train)
    return model
```

```python
def Random_forest():
    from sklearn.ensemble import RandomForestClassifier
    model = RandomForestClassifier()
    model.fit(x_train,y_train)
    return model
```

```python
def AdaBoost():
    from sklearn.ensemble import AdaBoostClassifier
    model = AdaBoostClassifier()
    model.fit(x_train,y_train)
    return model
```

```python
def CatBoost():
    from CatBoost import CatBoostClassifier
    model = CatBoostClassifier()
    model.fit(x_train,y_train)
    return model
```

```python
def XgBoost():
    from xgboost import XGBClassifier
    model = XGBClassifier()
    model.fit(x_train,y_train)
    return model
```

```python
models = []
models.append({'name': 'Decision_tree','model':Decision_tree()})
models.append({'name':'Random Forest','model': Random_forest()})
models.append({'name': 'AdaBoost','model':AdaBoost()})
models.append({'name':'CatBoost','model':CatBoost()})
models.append({'name':'XgBoost','model':XgBoost()})
```
