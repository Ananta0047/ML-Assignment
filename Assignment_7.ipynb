## Defining training and testing data

```python
x_train = df_train.drop('Salary',axis = 1)
y_train = df_train['Salary']
x_test = df_test.drop('Salary',axis = 1)
y_test = df_test['Salary']
```

# Model Building
- Logistic Regression
- Naive Bayes
- KNN
- SVM
- Decision Tree
- Random Forest
- Adaboost

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegressionCV
def create_LR():
    model = LogisticRegressionCV()
    model.fit(x_train,y_train)
    return model
```

#### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
def create_NB():
    model = GaussianNB()
    model.fit(x_train,y_train)
    return model
```

#### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
def create_KNN():
    model = KNeighborsClassifier()
    model.fit(x_train,y_train)
    return model
```

#### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
def create_DT():
    model = DecisionTreeClassifier()
    model.fit(x_train,y_train)
    return model
```

#### SVM

```python
from sklearn.svm import SVC
def create_svm():
    model = SVC()
    model.fit(x_train,y_train)
    return model
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
def create_RF():
    model = RandomForestClassifier()
    model.fit(x_train,y_train)
    return model
```

#### Adaboost

```python
from sklearn.ensemble import AdaBoostClassifier
def create_ADABoost():
    model = AdaBoostClassifier()
    model.fit(x_train,y_train)
    return model
```

```python
models = [
    {'name': 'Logistic Regression','model':create_LR()},
    {'name': 'Naive Bayes','model':create_NB()},
    {'name': 'KNN','model':create_KNN()},
    {'name': 'SVM','model':create_svm()},
    {'name': 'Decision Tree','model':create_DT()},
    {'name': 'Random Forest', 'model':create_RF()},
    {'name': 'Adaboost','model':create_ADABoost()}
] 
```

# Model Evaluation 

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

def model_evaluation(model):
        y_pred = model.predict(x_test)
        accuracy = accuracy_score(y_test,y_pred)
        precision = precision_score(y_test,y_pred)
        recall = recall_score(y_test,y_pred)
        f1 = f1_score(y_test,y_pred)
        cm = confusion_matrix(y_test,y_pred)
    
        return f"{accuracy:.2f}" , f"{precision:.2f}" , f"{recall:.2f}" , f"{f1:.2f}" 
```

```python
for model in models:
    accuracy , precision , recall , f1 = model_evaluation(model['model'])
    model['accuracy'] = accuracy
    model['precision'] = precision
    model['recall'] = recall
    model['f1'] = f1
    
```

```python
model_evaluation_report = pd.DataFrame(models,columns = ['name','accuracy','precision','recall','f1'])
model_evaluation_report
```
