Que1.-
Predict if a person would buy life insurance based on his age using logistic regression.
About Dataset - Age: Age of a person
bought_insurance : whether person buys Insurance or not (0 - No, 1 - Yes)
predict it for Age : 58,30,20,49,53

Dataset : insurance_data.csv

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot
import seaborn as sns
```

```python
df = pd.read_csv('insurance_data.csv')
```

### no need to clean data

```python
x = df.drop('bought_insurance',axis = True)
y = df['bought_insurance']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=123456)
```

```python
from sklearn.linear_model import LogisticRegressionCV
model = LogisticRegressionCV()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
type(y_pred)
```

```python
data = [58,30,20,49,53]
predictions = model.predict(pd.DataFrame(data))
```

```python
for i in range(len(predictions)):
    if i == 1:
        print(f"People with age {data[i]} will buy the insurance")
    else:
        print(f"People with age {data[i]} won't buy the insurance")
```
