# Removing outliers

```python
# Drop the 'Customer_Segment' column
df2 = df.drop('Customer_Segment', axis=1)

# Iterate through each column to calculate the IQR and filter out outliers
for i in df2.columns:
    # Calculate the 1st (Q1) and 3rd (Q3) quartiles
    Q1 = df2[i].quantile(0.25)
    Q3 = df2[i].quantile(0.75)
    IQR = Q3 - Q1
    
    # Calculate the lower and upper bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Filter rows that have values outside the bounds for the current column
    df2 = df2[(df2[i] >= lower_bound) & (df2[i] <= upper_bound)]

# After filtering, df2 will only contain rows where no column has outliers
df2.head(2)  # Print first 2 rows to check
```

```python
x = df.drop('Customer_Segment',axis = 1)
y = df['Customer_Segment']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=123456)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
y_pred = model.predict(x_test)
print(f"Accuracy: {accuracy_score(y_test,y_pred):.2f}, Precision: {precision_score(y_test,y_pred,average='macro'):.2f}, Recall: {recall_score(y_test,y_pred,average='macro'):.2f}")
```
